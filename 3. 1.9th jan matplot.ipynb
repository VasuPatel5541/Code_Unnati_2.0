#  Matplotlib

# 1. Graph

```python
import matplotlib.pyplot as plt
# x axis values
x = [1,2,3]
# y axis values
y = [2,4,1]
# plotting the points
plt.plot(x,y)
# naming the x axis
plt.xlabel('x-axis')
# naming the yaxis
plt.ylabel('y-axis')
plt.title("LINE GRAPH")
plt.show()
```

```python
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)

# figures in  matplotlib using figure()
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.show()
```

# 2. Subplotting 

```python
#plot-1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(2,2,1)
plt.plot(x,y)

#plot-2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(2,2,2)
plt.plot(x,y)

#plot-3
x = np.array([0,1,2,3])
y = np.array([11,28,19,24])
plt.subplot(2,2,3)
plt.plot(x,y)

#plot-4
x = np.array([0,1,2,3])
y = np.array([15,70,150,200])
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
# marker is used to mark the points on the graph
# marker will use all the shape to mark point
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
# marker is used to mark the points on the graph
# linestyle is used to plot the graph in dotted line
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dotted',marker='s')
plt.show()
```

```python
# MS is use to increase or decrease the size of marker
# MEC is used to color the border of the marker
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker='$Vasu$',ms=30,mec='r')
plt.show()
```

# SCATTER PLOT

```python
import matplotlib.pyplot as plt
# x axis values
x = [1,2,3,4,5]
# y axis values
y = [30,40,100,10,20]
# plotting the points
plt.scatter(x,y)
# naming the x axis
plt.xlabel('x-axis')
# naming the yaxis
plt.ylabel('y-axis')
plt.title("SCATTER GRAPH")
plt.show()
```

# LEGENDS , LABELS AND TITLES

```python
import matplotlib.pyplot as plt
import numpy as np
# x axis values
x = [5,10,15,20,25,30,35,40,45,50]
# y axis values
y = [1,4,3,2,7,6,9,8,10,5]
# plotting the points
plt.plot(x,y,'g')
# naming the x axis
plt.xlabel('x-axis')
# naming the yaxis
plt.ylabel('y-axis')
plt.tick_params(axis='x',labelsize=18,labelrotation=-60,labelcolor='black')
plt.tick_params(axis='y',labelsize=12,labelrotation=20,labelcolor='gray')
plt.show()
```

```python
import matplotlib.pyplot as plt
# x axis values
x = [1,2,3]
# y axis values
y = [2,4,1]
# plotting the points
plt.bar(x,y)
# naming the x axis
plt.xlabel('Time(hr)')
# naming the yaxis
plt.ylabel('Distance(km)')
plt.title("BAR GRAPH")
plt.show()
```

```python
#legend is used to mark the lines in graph
y1 = [2,3,4.5]
y2 = [1,1.5,5]
plt.plot(y1)
plt.plot(y2)
plt.legend(['blue','orange'],loc='best')
plt.show()
```

# HISTOGRAM

```python
data = np.random.normal(170,10,250)
#the hist() function will read the array and produce a histogram
plt.hist(data)
plt.show()

plt.hist(data,bins=30)
plt.show()
```

```python
#draw plot as fig
fig,ax = plt.subplots()
x = np.linspace(0,20,1000)

ax.plot(x, np.cos(x))
ax.axis('equal')
#text annotation to the plot where it indicate maximum value of the curw
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))
#text annotation to the plot where it indicate minimum value of the curv
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle='->',connectionstyle="angle3,angleA=0,angleB=-90"))
```

# LOGARITHMIC


```python
data = [10**i for i in range(5)]
plt.plot(data)
```

```python
datax = [10**i for i in range(5)]
datay = [i for i in range(5)]
plt.xscale("log")
plt.plot(datax,datay)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(1, 10, 100)
y = np.log10(x)  # Logarithmic function, base 10

# Plot the data with logarithmic axes
plt.plot(x, y, label='Logarithmic Function')

# Set logarithmic scale for both x and y axes
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.xlabel('X-axis (log scale)')
plt.ylabel('Y-axis (log scale)')
plt.title('Logarithmic Axes Example')

# Show the grid for better visualization
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.legend()
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig =plt.figure()
ax =plt.axes(projection='3d')

z = np.linspace(0, 1, 1000)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# plotting
ax.plot3D(x,y,z,'red')
ax.set_title('3D line plot')
plt.show()
```
